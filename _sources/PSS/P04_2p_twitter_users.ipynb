```python
fpath = "../_build/html/data/twitter_user1_hashed.csv"
import pandas as pd
df = pd.read_csv(fpath, on_bad_lines='skip') 
users = df.to_dict('records')
type(users)
```

## Counting

### Count by for-each

```python
from collections import Counter
lang_dict = Counter()
for user in users:
    lang_dict[user['account_language']] += 1

for k, v in lang_dict.most_common(10):
    print(k, v)
```

### Sampling: Traverse list by index and `range()`

如果你要做sampling，只想均勻地取出一半的資料出來做分析，那要怎麼做？

```python
lang_dict = Counter()
for i in range(len(users)):
    if i%2==0:
        lang_dict[users[i]['account_language']] += 1
for k, v in lang_dict.most_common(10):
    print("{}\t{}".format(k, v)) 
```

### Sampling: Traverse list by `enumerate()`

如果你要做sampling，只想均勻地取出一半的資料出來做分析，那要怎麼做？

```python
lang_dict = Counter()
for i, user in enumerate(users):
    if i%2==0:
        lang_dict[user['account_language']] += 1
for k, v in lang_dict.most_common(10):
    print("{}\t{}".format(k, v))
```
